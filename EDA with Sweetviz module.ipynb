> ### About Sweetviz
A pandas-based library to visualize and compare datasets.
Sweetviz is an open source Python library that generates beautiful, high-density visualizations to kickstart EDA (Exploratory Data Analysis) with a single line of code. Output is a fully self-contained HTML application.

>The system is built around quickly visualizing target values and comparing datasets. Its goal is to help quick analysis of target characteristics, training vs testing data, and other such data characterization tasks.

### Necessary imports

```python
import pandas as pd
import numpy as np
import sys
import os
import sweetviz as sv
# from pandas_profiling import ProfileReport
```

```python
### Dataset link: https://www.kaggle.com/c/house-prices-advanced-regression-techniques
import sweetviz
train = pd.read_csv("../data_sets/train.csv")
test = pd.read_csv("../data_sets/test.csv")
```

### Analyzing a single dataframe (and its optional target feature)

```python
my_report = sv.analyze([train, "Train"],target_feat='SalePrice')
```

```python
#the report will be generated in the form of .html type
my_report.show_html('Report.html')
```

```python


my_report.show_html('Report.html')
```

### Comparing two dataframes (e.g. Test vs Training sets)

```python

my_report1 = sweetviz.compare([train, "Train"], [test, "Test"], "SalePrice")
```

```python

my_report1.show_html('Report1.html')
```
